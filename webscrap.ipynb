```python
from textblob import TextBlob

def score_sentence_with_textblob(sentence):
    blob = TextBlob(sentence)
    polarity = blob.sentiment.polarity
    return polarity

# Example usage:
sentence = "fuck"
print(score_sentence_with_textblob(sentence))
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

sentence = "It was a good day but with some bad moments."
tokens = word_tokenize(sentence)
print(tokens)  # Output: ['It', 'was', 'a', 'good', 'day', 'but', 'with', 'some', 'bad', 'moments', '.']
```
