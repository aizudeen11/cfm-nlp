```python
def calc(P=1450, r=0.053, n=30, inc = 114):
    fv = 0
    for x in range(n):
        if fv > 0:
            ttl_month = 12
            P = P + inc
            fv2 = P*(((1+r/12)**ttl_month)-1)/(r/12)
            fv = fv + fv2
        elif fv == 0:
            ttl_month = 12
            fv2 = P*(((1+r/12)**ttl_month)-1)/(r/12)
            fv = fv + fv2
    return fv

calc()
```

```python
import pandas as pd
from prophet import Prophet
# Python
df1 = vix_history().rename({'Date':'ds', 'CLOSE':'y'}, axis=1)

df1 = df1.loc['2020-01-01':max(df1['ds'])]
df1.head()
```

```python
m = Prophet()
m.fit(df1)
```

```python
future = m.make_future_dataframe(periods=2, freq='ME')
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

```
