```python
from textblob import TextBlob

def score_sentence_with_textblob(sentence):
    blob = TextBlob(sentence)
    polarity = blob.sentiment.polarity
    return polarity

# Example usage:
sentence = "fuck"
print(score_sentence_with_textblob(sentence))
```

```python
def calc(P=1450, r=0.053, n=30, inc = 114):
    fv = 0
    for x in range(n):
        if fv > 0:
            ttl_month = 12
            P = P + inc
            fv2 = P*(((1+r/12)**ttl_month)-1)/(r/12)
            fv = fv + fv2
        elif fv == 0:
            ttl_month = 12
            fv2 = P*(((1+r/12)**ttl_month)-1)/(r/12)
            fv = fv + fv2
    return fv

calc()
```
